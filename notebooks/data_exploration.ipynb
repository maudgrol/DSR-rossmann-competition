# Data exploration Rossman data

## Set up

```python
# import relevant packages
import os
import pandas as pd
import numpy as np
```

```python
# set up paths
data_dir = "../data"
```

## Import and merge data

```python
train_df = pd.read_csv(os.path.join(data_dir, 'train.csv'))
store_df = pd.read_csv(os.path.join(data_dir, 'store.csv'))
```

```python
print(train_df.columns)
print(train_df.shape)
train_df
```

```python
print(store_df.columns)
print(store_df.shape)
store_df
```

```python
# Merge Store data with Train data
data = train_df.merge(store_df, how="left", left_on='Store', right_on='Store')
print(data.columns)
print(data.shape)
```

## Data cleaning

```python
data.loc[(data["Open"] == 1) & (data["Sales"].isna() )]
```

```python
data.loc[(data["Open"] == 0) & (data["Sales"] == 0 )]
```

### Drop rows with missing sales

```python
# check how many rows with missing or zero sales
print("There are {} rows with missing sales data" .format(data.loc[:, "Sales"].isnull().sum()))
print("There are {} rows with zero sales" .format((data["Sales"] == 0).sum()))
```

```python
# drop all rows with missing and zero sales (otherwise error in metric)
df = data.dropna(subset=["Sales"])
df = df.loc[~(df["Sales"] == 0)]
print(df.shape)
```

## Drop rows with missing store IDs

```python
df.isnull().sum()
```

```python
# Drop all rows with missing store IDs
df = df.dropna(subset=['Store'])
print(df.shape)
```

```python
# Missing values after dropping missing store IDs
df.isnull().sum()
```

### Check datatypes + change

```python
df.dtypes
```

```python
# Get relevant date features
df["Date"] = pd.to_datetime(df["Date"])
df["Month"] = df["Date"].dt.month
df["Year"] = df["Date"].dt.year
df["Weekday"] = df["Date"].dt.dayofweek 
df.drop(['DayOfWeek'], axis=1, inplace=True)
```

```python
df
```

### Save data file for further feature building

```python
df.to_csv(os.path.join(data_dir, 'df_clean.csv'), index=False)
```
